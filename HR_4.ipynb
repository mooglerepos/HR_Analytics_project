```python
#managng the null values
df.loc[df["NumCompaniesWorked"].isna(),"NumCompaniesWorked"] = 0
df.loc[df["NumCompaniesWorked"].isna()]
df['TotalWorkingYears'] = df['TotalWorkingYears'].fillna(0)

df.isna().sum()
```

```python
#dropping the columns
df.drop(['EmployeeCount', 'Over18','StandardHours', 'TrainingTimesLastYear', 'StockOptionLevel',
           'StandardHours', 'YearsWithCurrManager', 'YearsAtCompany'], axis=1, inplace=True)
df.head(1)
```

```python
# Import label encoder
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
```

```python
# Encode labels in columns
df['Attrition']= label_encoder.fit_transform(df['Attrition'])
df['Department']= label_encoder.fit_transform(df['Department'])
df['BusinessTravel']= label_encoder.fit_transform(df['BusinessTravel'])
df['EducationField']= label_encoder.fit_transform(df['EducationField'])
df['Gender']= label_encoder.fit_transform(df['Gender'])
df['MaritalStatus']= label_encoder.fit_transform(df['MaritalStatus'])
df['JobRole']= label_encoder.fit_transform(df['JobRole'])
df['City']= label_encoder.fit_transform(df['City'])
df['Business_type']= label_encoder.fit_transform(df['Business_type'])
df.head(20)
```

```python
#Split the data into independent 'X' and dependent 'Y' variables
X = df.drop("Attrition",axis=1)
Y = df['Attrition']
```

```python
# Split the dataset into 75% Training set and 25% Testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#Use Random Forest Classification algorithm
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
forest.fit(X_train, Y_train)
```

```python
#Get the accuracy on the training data
forest.score(X_train, Y_train)
```

```python
#Show the confusion matrix and accuracy for  the model on the test data
#Classification accuracy is the ratio of correct predictions to total predictions made.
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, forest.predict(X_test))
  
TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]
  
print(cm)
print('Model Testing Accuracy = "{}!"'.format(  (TP + TN) / (TP + TN + FN + FP)))
print()# Print a new line
```

```python
# Return the feature importances (the higher, the more important the feature).
importances = pd.DataFrame({'feature':X.iloc[:, 0:X.shape[1]].columns,'importance':np.round(forest.feature_importances_,3)}) #Note: The target column is at position 0
importances = importances.sort_values('importance',ascending=False).set_index('feature')
importances
```

```python
#Visualize the importance
importances.plot.bar()
```
